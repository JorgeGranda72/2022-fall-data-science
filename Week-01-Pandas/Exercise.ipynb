# Exercise 3 - Grouping

1. Using `groupby`, count how many listings are in each neighbourhood_group.


2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 


3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 


4. Using `groupby`, find the median price for each room type in each neighbourhood_group.


5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices for each room type in each neighbourhood_group.

```python
# 1. Using `groupby`, count how many listings are in each neighbourhood_group.

gb_nei = df.groupby('neighbourhood_group')['id'].count()
new_df = pd.DataFrame(gb_nei)
```

```python
new_df.reset_index()
```

```python
new_df = new_df.reset_index()
```

```python
# 2. Using `groupby`, find the mean price for each of the neighbourhood_groups. 

gb = df.groupby('neighbourhood_group')['price'].mean()
#gb['price'].agg( ['count', 'min', 'max', 'mean', 'median', 'std', 'sum'] )
mean_price = pd.DataFrame(gb)
```

```python
mean_price.reset_index()
```

```python
# 3. Using `groupby` and `.agg()`, find the min and max price for each of the neighbourhood_groups. 
gb2 = df.groupby('neighbourhood_group')['price'].max()
max_price = pd.DataFrame(gb2).reset_index()
max_price


```

```python
gb3 = df.groupby('neighbourhood_group')['price'].agg('min')
min_price = pd.DataFrame(gb3).reset_index()
min_price
```

```python
gb3 = df.groupby('neighbourhood_group')['price'].agg(['min','max'])
min_max_price = pd.DataFrame(gb3).reset_index()
min_max_price
```

```python
# 4. Using `groupby`, find the mean price for each room type in each neighbourhood_group.
gb4 = df.groupby(['neighbourhood_group','room_type'])['price'].agg('mean')

```

```python
df.columns
```

```python
gb6 = df.groupby(['neighbourhood_group','room_type'])
gb6['price'].agg('mean').reset_index()
```

```python
new_df = pd.DataFrame(gb4).reset_index()
new_df
```

```python
# Changing price column name
new_df.rename(columns={"price": "mean_price"},inplace = True)
new_df
```

```python
# 5. Using `groupby` and `.agg()`, find the count, min, max, mean, median, and std of the prices 
# for each room type in each neighbourhood_group.

gb6['price'].agg(['count', 'min', 'max', 'mean', 'median', 'std', 'sum']).reset_index()

```
